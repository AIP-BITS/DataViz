## Bar chart with Annotations

```python
x1=[1,3,5,7]
x2=[2,4,6,8]
y1=[7,7,7,7]
y2=[17,18,29,40]

plt.figure(figsize=(8,6))

bargraph1=plt.bar(x1,y1,label="First",color='green')
bargraph2=plt.bar(x2,y2,label="Second",color='red')

for bar1 in bargraph1:
    yval=bar1.get_height()
    plt.text(bar1.get_x()+bar1.get_width()/2,yval,int(yval),va='bottom')
    
for bar2 in bargraph2:
    yval=bar2.get_height()
    plt.text(bar2.get_x()+bar2.get_width()/2,yval,int(yval),va='bottom')
```

## Areachart

```python
time = np.arange(12)
income = np.array([5, 9, 6, 6, 10, 7, 6, 4, 4, 5, 6, 4])
expenses = np.array([6, 6, 8, 3, 6, 9, 7, 8, 6, 6, 4, 8])

fig, ax = plt.subplots(figsize=(8, 8))

# Plot lines
ax.plot(time, income, color="green")
ax.plot(time, expenses, color="red")

ax.fill_between(
    time, income,color="green", alpha=0.25
)

ax.fill_between(
    time, expenses,color="red", alpha=0.25
)
```

```python
time = np.arange(12)
income = np.array([5, 9, 6, 6, 10, 7, 6, 4, 4, 5, 6, 4])
expenses = np.array([6, 6, 8, 3, 6, 9, 7, 8, 6, 6, 4, 8])

fig, ax = plt.subplots(figsize=(8, 8))

# Plot lines
ax.plot(time, income, color="green")
ax.plot(time, expenses, color="red")

ax.fill_between(
    time, income,expenses,color="green", where=(income > expenses),alpha=0.25,interpolate=True
)

ax.fill_between(
    time, income,expenses,color="red", where=(income <= expenses),alpha=0.25,interpolate=True
)
```

```python
time = np.arange(12)
income = np.array([5, 9, 6, 6, 10, 7, 6, 4, 4, 5, 6, 4])
expenses = np.array([6, 6, 8, 3, 6, 9, 7, 8, 6, 6, 4, 8])

fig, ax = plt.subplots(figsize=(8, 8))

# Plot lines
ax.plot(time, income, color="green")
ax.plot(time, expenses, color="red")

# Fill area when income > expenses with green
ax.fill_between(
    time, income, expenses, where=(income > expenses), 
    interpolate=True, color="green", alpha=0.25, 
    label="Positive"
)

# Fill area when income <= expenses with red
ax.fill_between(
    time, income, expenses, where=(income <= expenses), 
    interpolate=True, color="red", alpha=0.25,
    label="Negative"
)

ax.legend();
```

# Stack Plot

```python
import matplotlib.pyplot as plt
 
# List of Days
days = [1, 2, 3, 4, 5]
 
# No of Study Hours
Studying = [7, 8, 6, 11, 7]
 
# No of Playing Hours
playing =  [8, 5, 7, 8, 13]
 
# Stackplot with X, Y, colors value
plt.stackplot(days, Studying, playing,
              colors =['r', 'c'])
 
# Days
plt.xlabel('Days')
 
# No of hours
plt.ylabel('No of Hours')
 
# Title of Graph
plt.title('Representation of Study and \
Playing wrt to Days')
 
# Displaying Graph
plt.show()
```

# Pie Chart

```python
import pandas as pd

sector = ["A","B","C","D","E"]
percentage = [20,35,25,5,15]

data = pd.DataFrame({"Sector":sector,"Percentage":percentage})
data
```

```python
fig,ax = plt.subplots()
ax.pie(data["Percentage"],labels=data['Sector'],autopct='%.2f')
plt.show()
```

```python
explode = (0, 0.1, 0, 0.3, 0)

fig,ax = plt.subplots()
ax.pie(data["Percentage"],labels=data['Sector'],autopct='%.2f',explode=explode)
plt.show()
```

```python
wedges, texts, autotexts = plt.pie(data['Percentage'], 
                                    labels=data['Sector'],
                                    autopct='%.2f',
                                    explode=explode)

plt.axis('equal')

wedges[1].set(edgecolor='blue', linewidth=2)

texts[1].set(family='cursive', size=20)

autotexts[1].set(weight='bold', size=15)
```

```python
colors = ['blue', 'grey', 'lawngreen']
bp = plt.boxplot(Z, 
                 patch_artist=True)

for i in range(len(bp['boxes'])):
    
    bp['boxes'][i].set(facecolor=colors[i])
    
    bp['caps'][2*i + 1].set(color=colors[i])

plt.xticks([1, 2, 3], ['A', 'B', 'C'])
```

```python
vp = plt.violinplot(Z,
                    showmedians=True,
                    vert=False)

plt.yticks([1, 2, 3], ['A', 'B', 'C'])

for i in range(len(vp['bodies'])):
    vp['bodies'][i].set(facecolor=colors[i]) 
    
vp['cmaxes'].set(color='maroon')

vp['cmins'].set(color='black')

vp['cbars'].set(linestyle=':')

vp['cmedians'].set(linewidth=6)

plt.legend(handles = [vp['bodies'][0], vp['bodies'][1],vp['bodies'][2]],
           labels = ['A', 'B','C'], 
           loc = 'upper left',
           bbox_to_anchor = (1,1)
           )
```

# twinx

```python
import numpy as np
import matplotlib.pyplot as plt

# Create some mock data
t = np.arange(0.01, 10.0, 0.01)
data1 = np.exp(t)
data2 = np.sin(2 * np.pi * t)

fig, ax1 = plt.subplots()

color = 'red'
# ax1.set_xlabel('time (s)')
# ax1.set_ylabel('exp', color=color)
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'blue'
# ax2.set_ylabel('sin', color=color)  # we already handled the x-label with ax1
ax2.plot(t, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
