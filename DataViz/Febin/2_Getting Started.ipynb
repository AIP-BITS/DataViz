### Matplotlib

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

### Create a simple plot

```python
plt.plot([2,4,6,8],
        [4,8,12,16])
```

```python
plt.plot([2,4,6,8],
        [4,8,12,16],
        color='red')
```

```python
plt.plot([2,4,6,8],
        [4,8,12,16],
        color='red')
plt.xlabel('x',fontsize=15,color='green')
plt.ylabel('2*x',fontsize=15,color='green')
```

```python
plt.plot([4,8,12,16])
```

### Sinewave

```python
x=np.linspace(start=0,stop=10,num=50)
```

```python
plt.plot(x,np.sin(x))
plt.xlabel('x',fontsize=15,color='green')
plt.ylabel('Sin(x)',fontsize=15,color='green')
```

### Formatting the ticks and tick labels

```python
plt.tick_params(axis='y',
               color='red',
               labelcolor='blue',
               labelsize='xx-large')
```

```python
plt.tick_params(axis='x',
               bottom=False,
               labelbottom=False)
```

```python
plt.plot([2,4,6,8],
        [4,8,12,16])
plt.ylim(ymin=0)
plt.xlim(xmin=0)
```

#### Adding legend

```python
plt.plot(x,np.sin(x),label='sine curve')
plt.legend()
```

```python
plt.plot(x,np.sin(x),label='sine curve')
plt.legend()
plt.title('Playing with plots')
```

```python
plt.plot(x,np.sin(x),label='sine curve')
plt.legend()
plt.title('Playing with plots')
plt.xlim(1,5)
```

```python
plt.plot(x,np.sin(x),label='sine curve')
plt.legend()
plt.title('Playing with plots')
plt.xlim(1,5)
plt.ylim(-1,0.5)
```

#### Inverting the axis

```python
plt.plot(x,np.sin(x),label='sine curve')
plt.legend(loc='upper center')
plt.title('Playing with plots')
plt.xlim(5,1)
plt.ylim(-1,0.5)
```

```python
plt.plot(x,np.sin(x),label='sine curve')
plt.legend(loc='upper center')
plt.title('Playing with plots')
plt.xlim(5,1)
plt.ylim(0.5,-1)
```
