```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(start=0,stop=10,num=50)
```

```python
plt.plot(x,np.sin(x))
plt.show()
```

### Multiple plots


```python
plt.plot(x,np.sin(x),label='sine curve',color='red')
plt.plot(x,np.cos(x),label='cosine curve',color='m')

plt.legend()
plt.title('Playing with plots')
```

#### plotting random array

```python
random_array=np.random.randn(20)
plt.plot(random_array,
        color='red')
```

### Linestyles

```python
plt.plot(random_array,
        color='red',
        linestyle=':')
```

```python
plt.plot(random_array,
        color='red',
        linestyle='--',
        linewidth=3)
```

```python
plt.plot(random_array,
        color='red',
         marker='d',
         markersize=10
        )
```

```python
plt.plot(random_array,
        color='red',
         marker='d',
         linestyle='None')
```

#### Scatterplot

```python
plt.scatter(range(0,20),
            random_array,
           color='red',
           marker='d')
```

#### Adding text to a plot

```python
xs=[2,4,6,8,10]
ys=[4,8,12,16,20]
plt.plot(xs,ys)

plt.text(2,4,'This text starts at point (2,4)')
plt.text(8,16,'This text ends at point (8,16)',horizontalalignment='right')

```

```python
plt.plot(xs,ys,'bo-')
plt.grid(True)

for x,y in zip(xs,ys):
    label= "{:.2f}".format(y)
    
    plt.annotate(label,
                (x,y),
                textcoords="offset points",
                xytext=(0,10),
                ha='center')
    
```

```python
plt.scatter(xs,ys)

for x,y in zip(xs,ys):
    label=f"({x},{y})"
    
    plt.annotate(label,
                (x,y),
                textcoords="offset points",
                xytext =(0,15),
                ha='center')
```
