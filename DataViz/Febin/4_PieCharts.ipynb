```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

### Load a dataset


```python
data=pd.read_csv('sector_weighting.csv')
data
```

### Pie Chart


```python
plt.pie(data['Percentage'],
       labels=data['Sector'])
```

#### Set the aspect of each axis to be equal


```python
plt.pie(data['Percentage'],
       labels=data['Sector'])

plt.axis('equal')
```

#### Pie chart customizations


```python
colors=['deeppink','aqua','magenta','silver','lime']
```

#### Draw the pie chart with the new colour palette and show percentage values


```python
plt.pie(data['Percentage'],
       labels=data['Sector'],
       colors=colors,
       autopct='%.2f')

plt.axis('equal')
```

#### Re-orienting the pie chart sectors


```python
plt.pie(data['Percentage'],
       labels=data['Sector'],
       colors=colors,
       startangle=180,
       counterclock=False)

plt.axis('equal')
```

#### The explode property


```python
explode=(0,0.1,0,0.3,0)
```

```python
plt.pie(data['Percentage'],
       labels=data['Sector'],
       explode=explode)

plt.axis('equal')
```

### Pie chart components


```python
wedges ,texts , autotexts = plt.pie(data['Percentage'],
       labels=data['Sector'],
        colors=colors,
        autopct='%.2f',
        explode=explode)
       
print('Wedges:', wedges)
print('Texts:', texts)
print('Autotexts:', autotexts)

plt.axis('equal')
```

#### Formatting the individual components of the pie chart



```python
wedges ,texts , autotexts = plt.pie(data['Percentage'],
       labels=data['Sector'],
        colors=colors,
        autopct='%.2f',
        explode=explode)
       
plt.axis('equal')

wedges[1].set(edgecolor='blue',linewidth=4)
texts[1].set(family='cursive',size=20)
```
