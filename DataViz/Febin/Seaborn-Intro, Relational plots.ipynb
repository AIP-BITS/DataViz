# Relational Plots

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.relplot(x="total_bill",y="tip",data=tips)
```

```python
sns.relplot(x="total_bill",y="tip",hue="smoker",data=tips)
```

```python
sns.relplot(x="total_bill",y="tip",hue='smoker', style='smoker', data=tips)
```

```python
sns.relplot(x="total_bill",y="tip",hue='smoker', style='time', data=tips)
```

```python
#sns.relplot(x="total_bill",y="tip",hue='size', style='smoker', data=tips)
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="size")
```

```python
sns.relplot(x="total_bill",y="tip",size="size", data=tips)
```

```python
sns.relplot(x="total_bill",y="tip",size="size",sizes=(15,200), data=tips)
```

```python
stock=pd.read_excel("STOCK.xlsx")
stock.head()
```

```python
plt.plot(stock['Date'],stock['Open'])
```

```python
for group_index,group_frame in stock.groupby('Company'):
    plt.plot(group_frame['Date'],group_frame['Open'],label=group_index)
plt.ylabel('Open price')
plt.legend()
```

```python
sns.lineplot(x='Date',y='Open',hue='Company',data=stock)
```

```python
stock_wide=stock.pivot(index='Date',columns='Company',values='Open')
stock_wide.head()
```

```python
for col in stock_wide.columns:
    stock_wide[col].plot()

plt.ylabel('Open price')
plt.legend()
```

```python
sns.relplot(data=stock_wide,kind='line')
```
