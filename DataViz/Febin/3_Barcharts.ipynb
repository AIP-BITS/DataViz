### Multiple Bars

```python
x1=[1,3,5,7]
x2=[2,4,6,8]
y1=[8,8,8,8]
y2=[17,18,29,40]

plt.figure(figsize=(8,6))

bargraph1=plt.bar(x1,y1,label='First',color='green')
bargraph2=plt.bar(x2,y2,label='Second',color='red')

for bar1 in bargraph1:
    yval = bar1.get_height()
    plt.text(bar1.get_x()+bar1.get_width()/2,yval,int(yval),va='bottom')
    
for bar2 in bargraph2:
    yval = bar2.get_height()
    plt.text(bar2.get_x()+bar2.get_width()/2,yval,int(yval),va='bottom')
    
plt.title('Bar Charts')
plt.legend()
```

```python


X=[28,33,43,45,5]
#X.sort()
Y=["C","B","A","D","E"]
y,x = zip(*sorted(zip(Y,X)))
# The above sort works:the inputs are zipped and sorted as you request elements from it.
#It is sorted as per Y values-Corresponding X values are picked.
#Example here A is the thris element in Y.So value for A is 43

plt.barh(y,x)
ax=plt.subplot()
ax.invert_yaxis()
```

```python
year1=(20,35,30,35,28,28,33)
year2=(25,39,45,2,21,29,36)

x=np.arange(7)
width=0.35

fig,ax=plt.subplots(figsize=(9,6))

rects=ax.bar(x-width/2,year1,width,color='blue',alpha=0.5,label='year1')
rects=ax.bar(x+width/2,year2,width,color='red',alpha=0.5,label='year2')

plt.legend()
```
