```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Dataset

```python
stock_data=pd.read_csv('stocks.csv')
stock_data.head()
```

```python
type(stock_data['Date'])
```

```python
stock_data['Date']=pd.to_datetime(stock_data['Date'])
stock_data.head()
```

```python
fig=plt.figure(figsize=(10,6))

ax1= fig.add_axes([0,0,1,1])
ax2= fig.add_axes([0.05,0.65,0.5,0.3])
```

```python
fig=plt.figure(figsize=(10,6))

ax1= fig.add_axes([0,0,1,1])
ax2= fig.add_axes([0.05,0.65,0.5,0.3])

ax1.plot(stock_data['Date'],
        stock_data['AAPL'],
        color='red')

ax1.set_title('AAPL vs IBM(inset)')

ax2.plot(stock_data['Date'],
        stock_data['IBM'],
        color='blue')
```

#### Compare multiple stocks

```python
fig=plt.figure(figsize=(10,6))
fig.suptitle('Stock price comparison 2007-2017',fontsize=20)

ax1 =fig.add_subplot(221)
ax1.set_title('MSFT')
ax1.plot(stock_data['Date'],
        stock_data['MSFT'],
        color='red')
```

```python
fig=plt.figure(figsize=(10,6))
fig.suptitle('Stock price comparison 2007-2017',fontsize=20)

ax1 =fig.add_subplot(221)
ax1.set_title('MSFT')
ax1.plot(stock_data['Date'],
        stock_data['MSFT'],
        color='red')

ax2 =fig.add_subplot(222)
ax2.set_title('GOOG')
ax2.plot(stock_data['Date'],
        stock_data['GOOG'],
        color='green')

ax3 =fig.add_subplot(223)
ax3.set_title('SBUX')
ax3.plot(stock_data['Date'],
        stock_data['SBUX'],
        color='m')

ax4 =fig.add_subplot(224)
ax4.set_title('IBM')
ax4.plot(stock_data['Date'],
        stock_data['IBM'],
        color='purple')
```
