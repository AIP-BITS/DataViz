## Plotting with categorical data

### Categorical scatterplots:

stripplot() (with kind="strip"; the default)

swarmplot() (with kind="swarm")

### Categorical distribution plots:

boxplot() (with kind="box")

violinplot() (with kind="violin")

boxenplot() (with kind="boxen")

### Categorical estimate plots:

pointplot() (with kind="point")

barplot() (with kind="bar")

countplot() (with kind="count")

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
```

### Categorical scatterplots


```python

tips=sns.load_dataset("tips")
sns.catplot(x="day",y="total_bill",data=tips)
```

```python

sns.catplot(x="day",y="total_bill",kind="swarm",data=tips)
```

```python
sns.catplot(x="day",y="total_bill",kind="swarm",hue="sex",data=tips)
```

## Distributions of observations within categories


#### Boxplots


```python
sns.catplot(x="day",y="total_bill",kind="box",hue="smoker",data=tips)
```

```python
tips["weekend"]=tips["day"].isin(["Sat","Sun"])
sns.catplot(x="day",y="total_bill",hue="weekend",kind="box",dodge=False,data=tips)
```

## Statistical estimation within categories


#### Bar plots


```python
titanic=sns.load_dataset("titanic")
titanic
```

```python
sns.catplot(x="sex",y="survived",hue="class",kind="bar",data=titanic)
```

```python
sns.countplot(x="deck",data=titanic)
```

```python
iris=sns.load_dataset("iris")
sns.catplot(data=iris,orient="h",kind="box")
```

### Showing multiple relationships with facets



```python
sns.catplot(x="day",y="total_bill",hue="smoker",col="time",kind="swarm",data=tips)
```

```python
titanic
```

```python
g=sns.catplot(x="fare",y="survived",row="class",kind="box",orient="h",data=titanic.query("fare>0"))

g.set(xscale="log")
```
