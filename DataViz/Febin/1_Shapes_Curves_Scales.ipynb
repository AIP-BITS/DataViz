```python
fig,ax=plt.subplots()
ax.set_aspect(aspect='equal')

polygon=patches.Polygon([[0.1,0.1],
                        [0.2,0.8],
                        [0.5,0.7],
                        [0.8,0.1],
                        [0.4,0.3]],
                        fill=False)

ax.add_patch(polygon)
```

### Scales
#### Linear line

```python
y=np.random.uniform(low=0.0,high=1000,size=(1000,))
y.sort()

x=np.arange(len(y))

plt.plot(x,y)
plt.grid(True)
plt.yscale('linear')
```

#### Logarithmic scale

```python
plt.plot(x,y)
plt.grid(True)
plt.yscale('log')
```

#### Changing the base of the log scale

```python
plt.plot(x,y)
plt.grid(True)
plt.yscale('log',basey=2)
```

#### Using different logarithmic scales on each axis

```python
plt.plot(x,y)
plt.grid(True)
plt.yscale('log',basey=10)
plt.xscale('log',basex=2)
```

#### Use a different distribution with negative values


```python
y=100*np.random.normal(size=1000)
y.sort()

x=np.arange(len(y))
```
