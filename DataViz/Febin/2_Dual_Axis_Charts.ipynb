```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
weather_data=pd.read_csv('weather.csv')
weather_data.head()
```

```python
weather_data =weather_data[['Date','TempAvgF','WindAvgMPH']].head(30)
weather_data 
```

```python
fig,ax_tempF=plt.subplots()

fig.set_figwidth(12)
fig.set_figheight(6)

ax_tempF.set_xlabel('Date')

ax_tempF.tick_params(axis='x',
                    bottom=False,
                    labelbottom=False)

ax_tempF.set_ylabel('Temp(F)',
                   color='red',
                   size='x-large')

ax_tempF.tick_params(axis='y',
                    labelcolor='green',
                    labelsize='large')


ax_tempF.plot(weather_data ['Date'],
             weather_data ['TempAvgF'],
             color='red')
```

```python
fig,ax_tempF=plt.subplots()

fig.set_figwidth(12)
fig.set_figheight(6)

ax_tempF.set_xlabel('Date')

ax_tempF.tick_params(axis='x',
                    bottom=False,
                    labelbottom=False)

ax_tempF.set_ylabel('Temp(F)',
                   color='red',
                   size='x-large')

ax_tempF.tick_params(axis='y',
                    labelcolor='green',
                    labelsize='large')


ax_tempF.plot(weather_data ['Date'],
             weather_data ['TempAvgF'],
             color='red')

ax_wind=ax_tempF.twinx()

ax_wind.set_ylabel('Avg Wind Speed',
                   color='blue',
                   size='x-large')

ax_wind.tick_params(axis='y',
                    labelcolor='blue',
                    labelsize='large')


ax_wind.plot(weather_data ['Date'],
             weather_data ['WindAvgMPH'],
             color='blue')
```

### Twin(Dual) axis for the same data

```python
def fahrenheit2celsius(f):
    return (f-32)*5/9
```

```python
fig,ax_tempF=plt.subplots()

fig.set_figwidth(12)
fig.set_figheight(6)

ax_tempF.set_xlabel('Date')

ax_tempF.tick_params(axis='x',
                    bottom=False,
                    labelbottom=False)

ax_tempF.set_ylabel('Temp(F)',
                   color='red',
                   size='x-large')

ax_tempF.tick_params(axis='y',
                    labelcolor='green',
                    labelsize='large')


ax_tempF.plot(weather_data ['Date'],
             weather_data ['TempAvgF'],
             color='red')

ax_tempC=ax_tempF.twinx()

ymin,ymax=ax_tempF.get_ylim()

ax_tempC.set_ylim(fahrenheit2celsius(ymin),
                 fahrenheit2celsius(ymax))

ax_tempC.set_ylabel('Temp(C)',
                   color='blue',
                   size='x-large')

ax_wind.tick_params(axis='y',
                    labelcolor='blue',
                    labelsize='large')


```
