# Null treatment

```python
data = data.dropna(axis=0)
data.shape
```

# Outlier Treatmnet

```python
Q1 = data['sepal_length'].quantile(0.25)
Q3 = data['sepal_length'].quantile(0.75)
IQR = Q3-Q1
```

```python
data.shape
```

```python
data = data[~((data['sepal_length'] < (Q1 - 1.5*IQR)) | (data['sepal_length'] > (Q3 + 1.5*IQR)))]
```

```python
fig,ax = plt.subplots(2,2,figsize=(18,8))
sns.boxplot(data=data,x='species',y='sepal_length',ax=ax[0][0])
sns.boxplot(data=data,x='species',y='sepal_width',ax=ax[0][1])
sns.boxplot(data=data,x='species',y='petal_length',ax=ax[1][0])
sns.boxplot(data=data,x='species',y='petal_width',ax=ax[1][1])
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

%matplotlib inline
```

```python
data = pd.read_csv("/content/drive/MyDrive/Dataviz-Python/DataViz/data/iris.csv")
```

```python
data = data.dropna(axis=0)
data.shape
```

```python
from scipy import stats
```

```python
data.shape
```

```python
import numpy as np
data = data[np.abs(stats.zscore(data[['sepal_length']]))<=3]
```

```python
fig,ax = plt.subplots(2,2,figsize=(18,8))
sns.boxplot(data=data,x='species',y='sepal_length',ax=ax[0][0])
sns.boxplot(data=data,x='species',y='sepal_width',ax=ax[0][1])
sns.boxplot(data=data,x='species',y='petal_length',ax=ax[1][0])
sns.boxplot(data=data,x='species',y='petal_width',ax=ax[1][1])
```

```python
sns.pairplot(data=data,hue='species')
```
