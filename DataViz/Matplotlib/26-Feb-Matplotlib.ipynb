```python
X = np.linspace(0,10,100)
Y1,Y2 = np.sin(X) , np.cos(X)
fig,(ax1,ax2) = plt.subplots(2,1)
ax1.plot(X,Y1,color="C1")
ax2.plot(X,Y2,color="C0")
plt.show()
```

```python
X = np.linspace(0,10,100)
Y1,Y2 = np.sin(X) , np.cos(X)
fig,(ax1,ax2) = plt.subplots(1,2)
ax1.plot(X,Y1,color="C1")
ax2.plot(X,Y2,color="C0")
plt.show()
```

```python
X = np.linspace(0,10,100)
Y1,Y2 = np.sin(X) , np.cos(X)

fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(X,Y1,color="C1")

ax2 = fig.add_axes([0.4,0.5,0.8,0.4])
ax2.plot(X,Y2,color="C0")

plt.show()
```

# Label

```python
X = np.linspace(0,10,100)
Y1,Y2 = np.sin(X) , np.cos(X)
fig,(ax1,ax2) = plt.subplots(1,2)
ax1.plot(X,Y1,color="C1")
ax2.plot(X,Y2,color="C0")

fig.suptitle("Wave")
ax1.set_title("Sine Wave")
ax2.set_title("Cos Wave")

ax1.set_ylabel("ylable")
ax2.set_xlabel("xlable")

plt.show()

```

```python
X = np.linspace(0,10,100)
Y1,Y2 = np.sin(X) , np.cos(X)
fig, ax = plt.subplots(3,3)
ax[0,0].plot(X,Y1,color="C1")
ax[2,2].plot(X,Y2,color="C0")
plt.show()
```

# Gridspec

```python
fig = plt.figure()
gs = fig.add_gridspec(2,2)

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[1,0])
ax3 = fig.add_subplot(gs[:,1])
ax1.plot(X,Y1,color="C1")
ax3.plot(Y2,X,color="C0")
plt.show()
```

# Spines

```python
fig,ax = plt.subplots()
ax.spines["left"].set_color("red")
ax.spines["right"].set_color("None")
ax.spines["top"].set_color("None")
```

# changing scales

```python
from matplotlib import markers
X = np.linspace(0,30,200)
Y1,Y2 = np.sin(X) , np.cos(X)
fig,ax=plt.subplots()
ax.set_xscale("log")
ax.plot(X,Y1)
plt.plot()
```

# fill_between

```python
X = np.linspace(0,30,20000)
Y1,Y2 = np.sin(X) , np.cos(X)
fig,ax=plt.subplots()
ax.plot(X[:10000],Y1[:10000])
ax.plot(X[10000:],Y1[10000:],linewidth=5)
ax.fill_betweenx([0,1],[0],[4],color="C7",alpha=0.2)
plt.plot()
```

```python
# Scatter plot
```

```python
X = np.random.uniform(0,1,100)
Y = np.random.uniform(0,1,100)
```

```python
fig,ax = plt.subplots()
ax.scatter(X,Y)
ax.scatter(X,Y)
plt.show()
```

```python
fig,ax = plt.subplots()

X = np.arange(4)
Y = [12,34,31,20]

bar = ax.bar(X,Y,width=0.5)
plt.xticks(X,('A','B','C','D'))

for i in bar:
  yval = i.get_height()
  plt.text(i.get_x()+i.get_width()/2,yval,int(yval))



plt.show()
```

```python
X = np.arange(4)
Y1 = [12,34,31,20]
Y2 = [18,23,13,45]

width =0.35

fig,ax = plt.subplots(figsize=(9,6))
bar1 = ax.bar(X-width/2,Y1,width,color='g',alpha=0.5,label='Y1')
bar1 = ax.bar(X+width/2,Y2,width,color='g',alpha=0.8,label='Y2')

plt.xticks(X,('A','B','C','D'))

plt.legend()

plt.show()
```

# Histogram

```python
Z = np.random.normal(0,1,100)
fig,ax = plt.subplots()
ax.hist(Z,edgecolor="b",bins=10)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.hist(Z,edgecolor="b",bins=10)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.hist(Z,edgecolor="b",bins=10,cumulative=True)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.hist(Z,edgecolor="b",bins=10,range=(-1.5,1.5))
plt.show()
```

```python
Z2 = np.random.uniform(0,10,100)
fig,ax = plt.subplots()
ax.hist(Z,edgecolor="b",bins=10)
ax.hist(Z2)
plt.show()
```

```python
fig,ax = plt.subplots()

ax.hist(Z2,edgecolor="C0")
ax.hist(Z,edgecolor="C1",bins=10,alpha=0.3)
plt.show()
```
