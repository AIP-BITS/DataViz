# Area Chart

```python
time = np.arange(12)
income = np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expenses = np.array([6,6,8,3,6,9,7,8,6,6,4,8])

fig,ax = plt.subplots()

ax.plot(time,income,color="red",alpha=0.25)
ax.plot(time,expenses,color="green",alpha=0.25)

ax.fill_between(time,income,color="red",alpha=0.25)
ax.fill_between(time,expenses,color="green",alpha=0.25)
```

```python
time = np.arange(12)
income = np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expenses = np.array([6,6,8,3,6,9,7,8,6,6,4,8])

fig,ax = plt.subplots()

ax.plot(time,income,color="red",alpha=0.25)
ax.plot(time,expenses,color="green",alpha=0.25)

ax.fill_between(time,income,expenses,color="red",alpha=0.25)
ax.fill_between(time,expenses,expenses,color="green",alpha=0.25)
```

```python
time = np.arange(12)
income = np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expenses = np.array([6,6,8,3,6,9,7,8,6,6,4,8])

fig,ax = plt.subplots()

ax.plot(time,income,color="red",alpha=0.25)
ax.plot(time,expenses,color="green",alpha=0.25)

ax.fill_between(time,income,expenses,color="red",where=(income>expenses),alpha=0.25)
ax.fill_between(time,income,expenses,color="green",alpha=0.25)
```

```python
time = np.arange(12)
income = np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expenses = np.array([6,6,8,3,6,9,7,8,6,6,4,8])

fig,ax = plt.subplots()

ax.plot(time,income,color="green",alpha=0.25)
ax.plot(time,expenses,color="red",alpha=0.25)

ax.fill_between(time,income,expenses,color="green",where=(income>expenses),alpha=0.25,interpolate=True)
ax.fill_between(time,income,expenses,color="red",where=(income<=expenses),alpha=0.25,interpolate=True)
```

```python
time = np.arange(12)
income = np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expenses = np.array([6,6,8,3,6,9,7,8,6,6,4,8])

fig,ax = plt.subplots()

ax.plot(time,income,color="green",alpha=0.25)
ax.plot(time,expenses,color="red",alpha=0.25)

ax.fill_between(time,income,expenses,color="green",where=(income>expenses),alpha=0.25,interpolate=True,label="good")
ax.fill_between(time,income,expenses,color="red",where=(income<=expenses),alpha=0.25,interpolate=True,label="bad")

ax.legend()
```

# Pie Chart

```python
sector = list("ABCDE")
percentage = [20,35,25,5,15]

data = pd.DataFrame({"Sector":sector,"Percentage":percentage})
```

```python
data
```

```python
fig, ax = plt.subplots()
ax.pie(data["Percentage"],labels=data["Sector"],autopct='%.2f')
plt.show()
```

```python
explode = (0,0.1,0,0,0)
fig, ax = plt.subplots()
ax.pie(data["Percentage"],labels=data["Sector"],autopct='%.2f',explode=explode)
plt.show()
```

```python
fig, ax = plt.subplots()
w,t,a = ax.pie(data["Percentage"],labels=data["Sector"],autopct='%.2f',explode=explode)
w[1].set(edgecolor='blue',linewidth=2)
t[1].set(size=20)
a[1].set(weight='bold')
plt.show()
```

# Box Plots

```python
Z = np.random.normal(10,1,(100,1))
fig,ax = plt.subplots()
ax.hist(Z)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.boxplot(Z)
plt.show()
```

```python
Z = np.append(Z,15)
Z = np.append(Z,16)
fig,ax = plt.subplots()
ax.boxplot(Z)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.boxplot(Z,vert=False)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.boxplot(Z,vert=False,notch=True)
plt.show()
```

```python
fig,ax = plt.subplots()
bp = ax.boxplot(Z,patch_artist=True,vert=False,notch=True)
bp['boxes'][0].set(facecolor='lightyellow',edgecolor="brown",hatch=".")
bp['medians'][0].set(linestyle='--')
bp['caps'][1].set(color='red')
plt.show()
```

# Violin Plot

```python
vp = plt.violinplot(Z)
plt.show()
```

```python
vp = plt.violinplot(Z,showmedians=True,vert=False,showmeans=True)

for i in range(len(vp['bodies'])):
  vp['bodies'][i].set(facecolor=colors[i])


plt.legend(handles = [vp["bodies"][0],vp['bodies'][1],vp['bodies'][2]],
           labels = ["blue","grey","pink"],
           loc="upper left",
           bbox_to_anchor=(1,1)
           
           )

plt.show()
```
