```python
import numpy as np 
```

```python
x = np.linspace(start=0,stop=10,num=50)
```

```python
plt.plot(x,np.sin(x))
plt.xlabel('x',fontsize=15,color='red')
plt.ylabel('sin(x)',fontsize=15,color='red')
```

```python
plt.plot(x,np.sin(x))
plt.tick_params(axis='y',
                color='red',
                labelcolor='blue',
                labelsize='xx-large'
                )
```

```python
plt.plot(x,np.sin(x))
plt.tick_params(axis='x',
                bottom=False,
                labelbottom=False  
                )
```

```python
plt.plot([2,4,6,8],
        [4,8,12,16],
         color='red')
plt.ylim(ymin=0)
plt.xlim(xmin=0)
plt.xlabel('x',fontsize=15,color='blue')
plt.ylabel('2x',fontsize=15,color='red')
```

```python
plt.plot(x,np.sin(x),label='Sine Curve1')
plt.legend()
plt.title('Sine Curve2')
```

```python
plt.plot(x,np.sin(x),label='Sine Curve1')
plt.legend()
plt.title('Sine Curve2')
plt.xlim(1,5)
```

```python
plt.plot(x,np.sin(x),label='Sine Curve1')
plt.legend()
plt.title('Sine Curve2')
plt.xlim(1,5)
plt.ylim(-1,0.5)
```

```python
plt.plot(x,np.sin(x),label='Sine Curve1')
plt.legend(loc='upper center')
plt.title('Sine Curve2')
plt.xlim(1,5)
plt.ylim(-1,0.5)
plt.savefig("/content/sine_curve.jpg")
```

```python
x = [2,4,6,8]
y = [4,8,12,16]
for x1,y1 in zip(x,y):
  print(x1,y1)
```

```python
x = [2,4,6,8]
y = [4,8,12,16]
plt.grid(True)
plt.scatter(x,y)
for x1,y1 in zip(x,y):
  lable = "({0},{1})".format(x1,y1)
  plt.annotate(lable,(x1,y1))
```
